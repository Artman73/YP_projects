```python
#–ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Ç–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö, –¥–æ–±–∞–≤–∏–º —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫, –≤—Ä–µ–º—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Å–µ—Ä–≤–∏—Å–∞(–¥–Ω–∏ –∏ –º–µ—Å—è—Ü—ã), —É–¥–∞–ª–∏–º –Ω–µ–∏–Ω—Ñ–æ—Ä–º —Å—Ç–æ–ª–±—Ü—ã
full_df['TotalCharges'] = pd.to_numeric(full_df['TotalCharges'], errors='coerce')

full_df['target'] = full_df['EndDate']
full_df.loc[full_df['EndDate'] != 'No', 'target'] = 1
full_df.loc[full_df['EndDate'] == 'No', 'target'] = 0

full_df['BeginDate'] = pd.to_datetime(full_df['BeginDate'], format='%Y-%m-%d')
full_df['EndDate'] = pd.to_datetime(full_df['EndDate'], errors='coerce', format='%Y-%m-%d %H:%M:%S')

full_df['duration_service_d'] = (full_df['EndDate'] - full_df['BeginDate']).dt.days
full_df['duration_service_m'] = round((full_df['EndDate'] - full_df['BeginDate']).dt.days/30)

full_df['duration_service_d'] = full_df['duration_service_d'].fillna(
    (datetime.strptime("2020-02-01", "%Y-%m-%d") - full_df['BeginDate']).dt.days)
full_df['duration_service_m'] = full_df['duration_service_m'].fillna(
    round((datetime.strptime("2020-02-01", "%Y-%m-%d") - full_df['BeginDate']).dt.days/30))

full_df = full_df.drop(columns=['customerID'], axis=1)

full_df['target'] = pd.to_numeric(full_df['target'], errors='coerce')

full_df.info()
full_df
```

### RandomForestClassifier + class_weight='balanced' + GridSearchCV

```python
#%%time

#–æ–±—ä—è–≤–ª—è–µ–º –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä
model = RandomForestClassifier(random_state=290822, class_weight='balanced')

#—Å–æ–∑–¥–∞–µ–º —Å–ª–æ–≤–∞—Ä—å, –≤ –∫–æ—Ç–æ—Ä—ã–π –≤–ø–∏—Å—ã–≤–∞–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã, –∫–æ—Ç–æ—Ä—ã–µ –±—É–¥–µ–º –ø—Ä–æ–≥–æ–Ω—è—Ç—å GridSearch‚Äô–µ–º
parametrs = { 'n_estimators': range (10, 51, 10),
              'max_depth': range (1, 15, 2),
              'min_samples_leaf': range (1, 8),
              'min_samples_split': range (2, 10, 2),
            }

#–∑–∞–ø—É—Å–∫–∞–µ–º –æ–±—É—á–µ–Ω–∏–µ (n_jobs = -1, –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤—Å–µ –¶–ü –¥–ª—è –æ–±—É—á–µ–Ω–∏—è, –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 1 - –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å 1 –¶–ü)
grid = GridSearchCV(model, parametrs, cv=3, n_jobs = -1, scoring='roc_auc', verbose=3) # f1
grid.fit(features_train, target_train)

#–ß—Ç–æ–±—ã –Ω–µ –ø–µ—Ä–µ–ø–∏—Å—ã–≤–∞—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –≤—Ä—É—á–Ω—É—é –∏ –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ —Å–æ–∑–¥–∞—Ç—å –º–æ–¥–µ–ª—å —Å –Ω–∞–∏–ª—É—á—à–∏–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏.
model_RandomForestClassifier_GridSearchCV = grid.best_estimator_

#–≤—ã–≤–æ–¥–∏–º –ª—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã
grid.best_params_
```

### CatBoostClassifier + GridSearchCV + upsampling

#### –¢–µ—Ö–Ω–∏–∫–∞ —É–≤–µ–ª–∏—á–µ–Ω–∏—è –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏ –ø—Ä–∏ –¥–∏—Å–±–∞–ª–∞–Ω—Å–µ –∫–ª–∞—Å—Å–æ–≤: upsampling
- –≠—Ç–∞ —Ç–µ—Ö–Ω–∏–∫–∞ –ø–æ–∑–≤–æ–ª—è–µ—Ç —Å–¥–µ–ª–∞—Ç—å –æ–±—ä–µ–∫—Ç—ã —Ä–µ–¥–∫–æ–≥–æ –∫–ª–∞—Å—Å–∞ –Ω–µ —Ç–∞–∫–∏–º–∏ —Ä–µ–¥–∫–∏–º–∏ –∏ –≤–∞–∂–Ω—ã–µ –∑–∞–¥–∞—á–∏ –ø–æ–≤—Ç–æ—Ä—è—é—Ç—Å—è –ø–æ –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ä–∞–∑, —á—Ç–æ–±—ã –∏—Ö –ª–µ–≥—á–µ –∑–∞–ø–æ–º–Ω–∏—Ç—å –º–æ–¥–µ–ª–∏.

```python
#—Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è upsampling
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

#—Å—Ä–∞–≤–Ω—è–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ 0 –∏ 1, –¥–æ–º–Ω–æ–∂–∏–≤ 1 –Ω–∞ 3
features_upsampled, target_upsampled = upsample(features_train, target_train, 3)
```

```python
#–æ–±—ä—è–≤–∏–º –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä  
model = CatBoostClassifier(random_state=290822)

#—Å–æ–∑–¥–∞–µ–º —Å–ª–æ–≤–∞—Ä—å, –≤ –∫–æ—Ç–æ—Ä—ã–π –≤–ø–∏—Å—ã–≤–∞–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã, –∫–æ—Ç–æ—Ä—ã–µ –±—É–¥–µ–º –ø—Ä–æ–≥–æ–Ω—è—Ç—å GridSearch‚Äô–µ–º
parametrs = {'reg_lambda':[1,3,6],
             'learning_rate':[0.001, 0.01,0.1]                           
            }

#–∑–∞–ø—É—Å–∫–∞–µ–º –æ–±—É—á–µ–Ω–∏–µ (n_jobs = -1, –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤—Å–µ –¶–ü –¥–ª—è –æ–±—É—á–µ–Ω–∏—è, –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 1 - –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å 1 –¶–ü)
grid = GridSearchCV(model, parametrs, cv=3, n_jobs = -1, scoring='roc_auc', verbose=0)
grid.fit(features_upsampled, target_upsampled)

#–≤—ã–≤–æ–¥–∏–º –ª—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã
print(grid.best_params_)
print(grid.best_score_)

model = grid.best_estimator_
dump(model, 'CatBoostClassifier_upsampling.joblib')
```
